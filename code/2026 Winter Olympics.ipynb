```python
# converting columns to datetime
schedule["start_time"] = pd.to_datetime(schedule["start_time"], format="%H:%M:%S")
schedule["end_time"] = pd.to_datetime(schedule["end_time"], format="%H:%M:%S")

# calculating time delta in months
schedule["td_mins"] = schedule["end_time"].dt.to_period("min") - schedule["start_time"].dt.to_period("min")

#schedule["td_mins"] = (schedule["end_time"]- schedule["start_time"]).dt.total_seconds().astype(int)
```

```python
schedule
```

```python
schedule['minutes'] = schedule['td_mins'] / pd.Timedelta(minutes=1)
```

Timing of the Rest Day
The rest day is set for Tuesday, February 10, 2026. This timing allows athletes to regroup after the opening weekend and before the second half of the competition begins. It is a common practice in multi-day sporting events to include rest days to enhance athlete performance and spectator experience.
